```python
df2 = {
    "Student_ID": [101, 102, 103, 104, 105],
    "CGPA": [3.8, 3.2, None, 3.9, 2.7],
    "Major": ["CSE", "EEE", "Business", "CSE", "Economics"],
    "Internships": [2, 1, 0, 3, None],
    "Placed": ["Yes", "No", "No", "Yes", "No"]
}

df2 = pd.DataFrame(df2)

df2['CGPA'].fillna(df2['CGPA'].mean(), inplace=True)

df2['Internships'].fillna(0, inplace=True)

display(df2)
```

```python
cse_students = df2[df2['Major'] == 'CSE']
cse = cse_students.nlargest(1, 'CGPA')

display(cse)
```

```python
df2_encoded = pd.get_dummies(df2, columns=['Major', 'Placed'])
display(df2_encoded)
```
